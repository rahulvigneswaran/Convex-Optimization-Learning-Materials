# Libraries Required

```
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Examples of Convex Functions

---



1.   Affine Function
2.   Exponential Function
3.   Power Function
4.   Powers of Absolution Value
5.   Negative Entropy





#### 1.   Affine Function

---




```
x = np.linspace(-10,10);
a = 20;
b = 10;

fig, ax = plt.subplots()
ax.plot(x, a*x+b)


ax.set(title='Affine Function')
ax.grid()
plt.show()
```


#### 2.   Exponential Function

---


```
x = np.linspace(-10,10);
a = 2;
b = 10;

fig, ax = plt.subplots()
ax.plot(x, pow(x,a))


ax.set(title='Exponential Function')
ax.grid()
plt.show()
```


#### 3.   Powers

---


This function is convex only if a>=1 or a<=0 . If a is between 0 and 1, then it becomes concave in nature. It should be noted that the entire function will be convex only in R<sub>++</sub>

<iframe src="https://www.desmos.com/calculator/kqhrqc00rm?embed" width="500px" height="500px" style="border: 1px solid #ccc" frameborder=0></iframe>

```
x = np.linspace(0,10)

a=3

plt.subplot(311)
plt.plot(x, pow(x,a))
plt.title('Power Function where a>=1')
plt.grid()

a=-1
plt.subplot(312)
plt.plot(x, pow(x,a))
plt.title('Power Function where a<=0')
plt.grid()
a=0.5

plt.subplot(313)
plt.plot(x, pow(x,a))
plt.title('Power Function where a is in between 0 and 1')

plt.tight_layout()
plt.grid()
plt.show()
```


#### 4.   Powers of absolute value

---
Only applicable for a>=1


```
x = np.linspace(-10,10)

a=3

plt.subplot(311)
plt.plot(x, pow(abs(x),a))
plt.title('Power Function where a>=1')

plt.tight_layout()
plt.grid()
plt.show()
```


#### 5.  Negative entropy

---
Only applicable in R<sub>++</sub>. In R<sub>- -</sub>, the log value doesnt exist.


```
x = np.linspace(0.0001,2,100000)

a=3

plt.subplot(311)

plt.plot(x, x*(np.log10(x)))
plt.title('Negative entropy')

plt.tight_layout()
plt.grid()
plt.show()
```
